```python
# Write a Python program to calculate compound interest. Use the formula: 
# Where: 
# A= total amount 
# P = principal amount 
# r = annual interest rate (decimal) 
# n = number of times interest is compounded per year 
# t = time in years 
# Take P, r, n, and t as inputs from the user.

p = float(input("Enter principle"))
r = float(input("Enter annual intrest rate(decimal):"))
n = float(input("Enter number of times interest is compounded per year"))
t = float(input("time in years"))
A = p*(1+r/n)**(n-t)
print("Toatal Amount", A)
```

```python
# Write a Python program to find the roots of a quadratic equation. Use the formula: 
# X = -b+-rediaclb2-4ac/2a
# Take a, b, and c as inputs from the user.

a = float(input("Enter value of a:"))
b = float(input("Enter value of b:"))
c = float(input("Enter value of c:"))
d = b*b-4*a*c
root_d = 0.5**d
root1 = (-b + root_d) / (2*a)
root2 = (-b - root_d) / (2*a)
print("Root1", root1)
print("Root2", root2)

```

```python
# Write a Python program to calculate the future value of an investment. Use the formula: 
# ğ¹ğ‘‰ = ğ‘ƒğ‘‰(1 + ğ‘Ÿ)
# where: 
# o FV= future value 
# o PV= present value 
# o r= annual interest rate (as a decimal) 
# o t= time in years 
# Take PV, r, and t as inputs from the user.

PV = float(input("Enter present value:"))
r = float(input("Enter annual interest rate (as a decimal):"))
t = float(input("Enter time in years :"))
FV = PV*(1+r)**t
print("Future value=",FV)
```

```python
# Write a Python program to calculate the work done by a force. Use the formula: 
# where: W = f.dcos@ 
# W = work done 
# f= force 
# d = distance 
# theta = angle between force and direction of movement (in degrees) 
# Take f, d, and Î¸ as inputs from the user

f = float(input("Enter force :"))
d = float(input("Enter distance:"))
theta  = float(input("Enter value of angel:"))
pi = 3.14
x = theta*(pi/180)
cosx = 1 - (x*x)/2 + (x**4)/24
W = f*d*cosx
print("Work done:",W)

```
